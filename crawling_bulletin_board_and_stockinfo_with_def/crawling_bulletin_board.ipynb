### Creating Custom Definition

```python
def board_st_scp(typ,item_code, page_no):
    # board_info    
    url = f'https://finance.naver.com/item/{typ}.nhn?code={item_code}&page={page_no}'
    headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36"}
    response = requests.get(url, headers = headers)
    html = bs(response.text)

    tables = html.select("table")
    
    bd_info = pd.read_html(str(tables), encoding = 'cp949')
    return bd_info[0].dropna(how='all') 
```

```python
def board_st_scp(typ,item_code, page_no):
    # board_info    
    url = f'https://finance.naver.com/item/{typ}.nhn?code={item_code}&page={page_no}'
    headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36"}
    response = requests.get(url, headers = headers)
    html = bs(response.text)

    tables = html.select("table")
    
    bd_info = pd.read_html(str(tables), encoding = 'cp949')
    if typ == 'sise_day':
        bd_info = bd_info[0].dropna(how='all')
    else:
        bd_info = bd_info[1].dropna(how='all')
    return bd_info.dropna(how='all') 
```

###### <font color="blue">생성함수 사용설명</font>
    - 만약 일별 주식정보를 불러오길 원한다면, board_st_scp('sis_day','주식번호', 페이지넘버)를 입력
    - 만약 종목토론 주제들을 불러오길 원단하면, board_st_scp('board','주식번호', 페이지넘버)를 입력

```python
# sise_day /  board
board_st_scp('board','039010', 2)
```
